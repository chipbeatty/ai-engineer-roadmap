```python
import pandas as pd
import numpy as np

# --- Data Loading ---
file_path = 'data/playlist_2010to2023.csv'

try:
    df = pd.read_csv(file_path, encoding='latin1')
    print("Dataset loaded successfully!")
except UnicodeDecodeError:
    print(f"UnicodeDecodeError with 'latin1' encoding for {file_path}. Trying 'cp1252'...")
    try:
        df = pd.read_csv(file_path, encoding='cp1252') # <--- Fallback to cp1252
        print("Dataset loaded successfully with 'cp1252' encoding!")
    except Exception as e:
        print(f"Failed to load with 'cp1252' either. Error: {e}")
        print("You might need to try other encodings or inspect the file directly.")
        import sys
        sys.exit(1)
except FileNotFoundError:
    print(f"Error: File not found at {file_path}. Please check the file name and path.")
    import sys
    sys.exit(1) # Exit the notebook execution
except Exception as e:
    print(f"An unexpected error occurred during loading: {e}")
    import sys
    sys.exit(1)

# --- Initial Inspection ---
print("\n--- First 5 rows of the dataset ---")
print(df.head())

print("\n--- Dataset Information (columns, non-null counts, data types) ---")
df.info()

print("\n--- Basic Descriptive Statistics for Numerical Columns ---")
print(df.describe())

print("\n--- Missing Values Count per Column ---")
print(df.isnull().sum())

# --- Check for duplicate rows (common initial check) ---
print("\n--- Number of Duplicate Rows ---")
print(df.duplicated().sum())
```

```python
# --- Data Cleaning ---

# Start with a copy of the original DataFrame to perform cleaning/transformations on
df_cleaned = df.copy()
print("Created a copy of the DataFrame for cleaning: df_cleaned.")

# Since you confirmed no missing values or duplicates in the original data,
# we can skip df.dropna() and df.drop_duplicates() for THIS dataset.
# In future projects, these lines would be crucial here!

# 3. Data Type Conversion (Example: Ensure 'year' is integer, if it exists and is object type)
# Some datasets might load 'year' as object if there are any non-numeric entries like "2010s"
# We'll include this as a general check, but if your 'year' is already int, it won't change anything.
if 'year' in df_cleaned.columns and df_cleaned['year'].dtype == 'object':
    # Try converting to numeric, errors='coerce' will turn non-convertible values into NaN
    df_cleaned['year'] = pd.to_numeric(df_cleaned['year'], errors='coerce')
    # If any NaNs were introduced by coerce (meaning non-numeric 'year' values), drop those rows
    df_cleaned.dropna(subset=['year'], inplace=True)
    df_cleaned['year'] = df_cleaned['year'].astype(int) # Convert to integer after cleaning
    print("\n'year' column converted to integer (if applicable).")
elif 'year' in df_cleaned.columns and df_cleaned['year'].dtype != 'int64':
    print(f"Warning: 'year' column has unexpected dtype: {df_cleaned['year'].dtype}. Expected int64 or object for conversion.")


# 4. Simple Feature Engineering: Convert duration from milliseconds to minutes
if 'duration_ms' in df_cleaned.columns:
    df_cleaned['duration_minutes'] = df_cleaned['duration_ms'] / 60000
    print("\n'duration_minutes' column added.")

# --- Verify Cleaned Data ---
print("\n--- Cleaned Dataset Information ---")
df_cleaned.info()
print("\n--- Cleaned Dataset Head ---")
print(df_cleaned.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for plots for better aesthetics
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Set a default figure size for better visibility

# --- Example 1: Distribution of a Numerical Feature (e.g., 'danceability') ---
# Histogram to show distribution
plt.figure(figsize=(8, 5)) # You can set specific figure sizes for each plot
sns.histplot(df_cleaned['danceability'], kde=True, bins=20)
plt.title('Distribution of Danceability in Spotify Top Hits')
plt.xlabel('Danceability')
plt.ylabel('Count')
plt.show()

# --- Example 2: Trends over Time (e.g., Average Track Popularity by Year) ---
# Ensure 'year' column exists and is suitable for grouping
if 'year' in df_cleaned.columns and pd.api.types.is_numeric_dtype(df_cleaned['year']):
    avg_popularity_by_year = df_cleaned.groupby('year')['track_popularity'].mean().reset_index()
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=avg_popularity_by_year, x='year', y='track_popularity', marker='o')
    plt.title('Average Track Popularity by Year')
    plt.xlabel('Year')
    plt.ylabel('Average Popularity')
    plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
    plt.grid(True)
    plt.show()
else:
    print("Skipping 'Average Track Popularity by Year' plot: 'year' column not found or not numeric.")


# --- Example 3: Top N Artists (Categorical Data) ---
# Ensure 'artist_name' column exists
if 'artist_name' in df_cleaned.columns:
    top_artists = df_cleaned['artist_name'].value_counts().head(10) # Get top 10 artists
    plt.figure(figsize=(12, 7))
    sns.barplot(x=top_artists.values, y=top_artists.index, hue=top_artists.index, palette='viridis', legend=False)
    plt.title('Top 10 Artists in Spotify Top Hits')
    plt.xlabel('Number of Songs')
    plt.ylabel('Artist Name')
    plt.show()
else:
    print("Skipping 'Top 10 Artists' plot: 'artist_name' column not found.")


# --- Example 4: Relationship between two numerical features (e.g., Danceability vs. Energy) ---
if 'danceability' in df_cleaned.columns and 'energy' in df_cleaned.columns:
    plt.figure(figsize=(8, 8))
    sns.scatterplot(data=df_cleaned, x='danceability', y='energy', alpha=0.5)
    plt.title('Danceability vs. Energy')
    plt.xlabel('Danceability')
    plt.ylabel('Energy')
    plt.show()
else:
    print("Skipping 'Danceability vs. Energy' plot: required columns not found.")
               
               
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_cleaned is already loaded. If not, you'd load it like:
# df_cleaned = pd.read_csv('your_dataset.csv') # Replace 'your_dataset.csv' with the actual path

# Set a style for plots for better aesthetics (usually done once at the start of your script)
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Set a default figure size for better visibility

# --- Plot 3: Average Loudness and Energy Trends by Year ---
if 'year' in df_cleaned.columns and pd.api.types.is_numeric_dtype(df_cleaned['year']) and \
   'loudness' in df_cleaned.columns and pd.api.types.is_numeric_dtype(df_cleaned['loudness']) and \
   'energy' in df_cleaned.columns and pd.api.types.is_numeric_dtype(df_cleaned['energy']):
    
    # 1. Data Preparation: Aggregate average loudness and energy by year
    avg_loudness_by_year = df_cleaned.groupby('year')['loudness'].mean().reset_index()
    avg_energy_by_year = df_cleaned.groupby('year')['energy'].mean().reset_index()

    # 2. Plot Setup: Create the figure and set its size
    plt.figure(figsize=(12, 6))

    # 3. Main Plotting Functions: Create two line plots on the same axes
    sns.lineplot(data=avg_loudness_by_year, x='year', y='loudness', marker='o', label='Average Loudness')
    sns.lineplot(data=avg_energy_by_year, x='year', y='energy', marker='o', label='Average Energy')

    # 4. Add Labels, Title, and Legend
    plt.title('Average Loudness and Energy of Spotify Top Hits by Year')
    plt.xlabel('Year')
    plt.ylabel('Average Feature Value') # A general label as both loudness and energy are on this axis
    plt.grid(True)                     # Add a grid for readability
    plt.legend()                       # Display the legend to distinguish lines

    # 5. Display the Plot
    plt.show()
else:
    print("Skipping 'Average Loudness and Energy by Year' plot: required columns not found or not numeric.")
```
